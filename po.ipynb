```python
import pandas as pd
df= pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
# Get summary statistics of the dataset to understand the distribution of numerical features and identify any potential outliers.
df.describe()
```

```python
# To get a comprehensive overview of the dataset, including both numerical and categorical features, we can use the `include='all'` parameter in the `describe()` method. This will provide summary statistics for all columns, including counts, unique values, top values, and frequency for categorical features.
df.describe(include='all')
```

```python
# To understand the structure of the dataset, including the number of rows, columns, data types, and non-null counts, we can use the `info()` method. This will help us identify any missing values and understand the types of data we are working with.
df.info()
```

```python
# Check for missing values in each column to identify any data quality issues that need to be addressed before analysis.
df.isnull().sum()
```

```python
# Replace missing values in the 'Review Rating' column with the median rating for each product category to maintain the integrity of the data while handling missing values.
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Standardize column names by converting them to lowercase and replacing spaces with underscores to ensure consistency and ease of use in subsequent analysis.  
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
# create a new columns for age group
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels= labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create new column purchase_frequency_days 
frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
```

```python
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```
